# Problema 1

Para hacer un programa que imprima los números del -100 al 200, se escribe el código for y se especifíca el rango deseado a monstrar con la palabra clave. Se pone hasta 201, pues se cuenta desde el 0.

```python

for num in range(-100, 201):
  print(num)

```

# Problema 2

Para hacer una lista de números del -100 al 200 divisibles entre 6, volvemos a utilizar num, con el rango que ya le habíamos especificado antes, pero ahora agregamos la condición con 'if' para verificar si cada número del rango era divisible entre 6 utilizando %, que calcula el resto de la división.

```python
for num in range(-100, 201):
  if num % 6 == 0:
    print(num)
```

# Problema 3

Aquí, necesitábamos leer dos números, sumarlos y luego comparar la suma con ciertos valores. Siguiendo el pseudocódigo proporcionado, leí los dos números, los sumé y luego utilicé una serie de declaraciones if y elif para determinar en qué rango caía la suma y asignar el mensaje correspondiente a la variable resultado. Finalmente, imprimí el resultado.

```python

numero1 = float(input("Ingrese el primer número: "))
numero2 = float(input("Ingrese el segundo número: "))
suma = numero1 + numero2

if suma < 200:
    resultado = "menor a 200"
elif 200 <= suma < 250:
    resultado = "mayor a 200 y menor a 250"
else:
    resultado = "mayor a 250"

print(resultado)
```

# Problema 4

En este problema, se nos pedía evaluar si el usuario era mayor de edad y si le gustaba la música urbana, y luego enviar diferentes textos según la respuesta del usuario. Leí la edad del usuario y su preferencia musical, luego utilicé una declaración if para verificar ambas condiciones. Si ambas condiciones se cumplían, asigné un mensaje específico a la variable resultado, de lo contrario, asigné otro mensaje. Luego, imprimí el resultado.

```python

edad = int(input("¿Cuál es tu edad? "))
gusta_musica_urbana = input("¿Te gusta la música urbana? (responde sí o no): ").lower()

if edad > 18 and gusta_musica_urbana == "sí":
    resultado = "Eres mayor de edad y te gusta la música urbana"
else:
    resultado = "No eres mayor de edad o no te gusta la música urbana"

print(resultado)
```
