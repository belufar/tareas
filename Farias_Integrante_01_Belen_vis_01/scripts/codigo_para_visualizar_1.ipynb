```python
import pandas as pd

# Leer el archivo CSV especificando el separador ';' y el separador decimal ','
file_path = '/content/drive/My Drive/GRAFICA/1306/databaseregiones.csv'
df = pd.read_csv(file_path, sep=';', encoding='utf-8', decimal=',')

# Crear una lista para almacenar los resultados
resultados = []

# Iterar sobre cada columna (excepto la primera que es "País")
for region in df.columns[1:]:
    # Encontrar el índice del país con el mayor porcentaje en la región
    max_idx = df[region].idxmax()
    # Obtener el país y el valor máximo
    max_pais = df.loc[max_idx, 'País']
    max_valor = df.loc[max_idx, region]
    # Almacenar el resultado en un diccionario
    resultado = {'Región': region, 'Nacionalidad con mayor %': max_pais, 'Porcentaje': max_valor}
    resultados.append(resultado)

# Crear un DataFrame con los resultados
df_resultados = pd.DataFrame(resultados)

print(df_resultados)
```

```python
import matplotlib.pyplot as plt


# Crear el gráfico de barras apiladas
fig, ax = plt.subplots(figsize=(12, 8))

# Obtener los valores y etiquetas para el gráfico
nacionalidades = df_resultados['Nacionalidad con mayor %']
porcentajes = df_resultados['Porcentaje']
regiones = df_resultados['Región']

# Crear el gráfico de barras apiladas
ax.barh(regiones, porcentajes, color='skyblue')

# Añadir etiquetas y título
ax.set_xlabel('Porcentaje')
ax.set_ylabel('Región')
ax.set_title('Principal país de origen de migrantes en cada región de Chile')

# Añadir etiquetas a las barras
for i in range(len(regiones)):
    ax.text(porcentajes[i], i, f'{nacionalidades[i]} ({porcentajes[i]}%)', va='center')

# Mostrar el gráfico
plt.tight_layout()
plt.show()

# Guardar el gráfico como una imagen JPEG
plt.savefig('grafico.jpg', format='jpg')

```
